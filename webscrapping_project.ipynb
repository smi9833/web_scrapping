```python
import requests as re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.themoviedb.org/movie"
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
page=re.get(url,headers=header)
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup
```

```python
scrap_movie=soup.findAll('div', class_='card style_1')
```

```python
scrap_movie
```

```python
movies_link=[]
for movie in scrap_movie:
    m_link=movie.find('h2').a['href']
    movies_link.append('https://www.themoviedb.org/'+m_link)
```

```python
movies=[]
for links in movies_link:
    header1 = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
    page=re.get(links,headers=header1)
    page.content
    inner_soup=BeautifulSoup(page.content,'html.parser')
    inner_movie_link=('https://www.themoviedb.org/'+inner_soup.find('h2').a['href'])
    m_title=inner_soup.find('h2').text.strip().replace('\n',',')
    rel_date=inner_soup.find('span',class_='release').text.split().pop(0)
    genres=inner_soup.find('span', class_='genres').text.strip()
    runtime=inner_soup.find('span',class_='runtime').text.strip()
    rating=inner_soup.find('div',class_='user_score_chart')['data-percent']
    overview=inner_soup.find('div',class_='overview').p.text
    #director=inner_soup.find('li', class_='profile').p.text
    movie_dict={
        'Movie Name':m_title,
        'Movie Link':inner_movie_link,
        'Release Date':rel_date,
        'User Rating':rating,
        'Genres':genres,
        'Duration':runtime,
        'Overview':overview      
    }
    #print(m_title)
    movies.append(movie_dict)
    print(movies)
```

```python
movies_list=pd.DataFrame(movies)
movies_list
```

```python
movies_list.to_csv("Hollywood Movie List.csv")
```
